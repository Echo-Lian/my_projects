```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## 1 Websrcaping from web

```python
# initialize a list to store url
url = []

# read url from external file
with open("webpages.txt", "r") as file:
    for line in file:
        url.append(line)

# initialize a list to store the webpage
raw_data = []

# loop over every webpage and read to beautiful data
for link in url:
    html_data = requests.get(link)
    beautiful_soup = BeautifulSoup(html_data.content, 'html.parser')
    raw_data.append(beautiful_soup)
```
